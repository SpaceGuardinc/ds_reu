```python
"""Переименование столбцов"""
answer_df.rename(columns={'name_product':'Продукт',
                         'weight_product':'Вес',
                         'product_expiration_date':'Годен продукт до',
                         'calories_product':'Калории продукта',
                         'wholesale_price_product':'Оптовая цена',
                         'quantity':'Количество продукта',
                         'date_manufacture':'Дата производства',
                         'defective':'Брак',
                         'quantity_1':'количество ингридиектов',
                         'product_expiration_date_1':'Годен ингридиет до',
                         'calories_product_1':'Калории ингридиента',
                         'price':'Цена ингридиета',
                         'quantity_2':'Кол-во купленных ингридиетов',
                         'procurement_date':'Дата закупки',
                         'added':'Оформлен ли заказ'})
```
