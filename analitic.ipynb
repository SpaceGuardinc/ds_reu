```python
from database.server_methods.get import get_products
import matplotlib.pyplot as plt
```

```python
answer_df = get_products()
```

```python
print(answer_df)
```

```python
"""Переименование столбцов"""
answer_df.rename(columns={'name_product':'Продукт',
                         'weight_product':'Вес',
                         'product_expiration_date':'Годен продукт до',
                         'calories_product':'Калории продукта',
                         'wholesale_price_product':'Оптовая цена',
                         'quantity':'Количество продукта',
                         'date_manufacture':'Дата производства',
                         'defective':'Брак',
                         'quantity_1':'количество ингридиектов',
                         'product_expiration_date_1':'Годен ингридиет до',
                         'calories_product_1':'Калории ингридиента',
                         'price':'Цена ингридиета',
                         'quantity_2':'Кол-во купленных ингридиетов',
                         'procurement_date':'Дата закупки',
                         'added':'Оформлен ли заказ'})
```

```python
"""Сортировка продуктов по их количеству"""
answer_df.drop_duplicates("quantity").sort_values(by = 'name_product').head().plot(figsize=(15,3))
plt.legend(["Вес продукта", 
            "Калории продукта", 
            "Оптовая цена", 
            "Количество продутка", 
            "Брак продукта",
            "Количество ингридиентов",
            "Цена ингридиентов",
            "Количество заказов"])
```

```python
"""Общий вид данных в виде обычного графика"""
answer_df.drop_duplicates(subset=['name_product']).plot(figsize=(15,5)).legend(["Вес продукта", 
                        "Калории продукта", 
                        "Оптовая цена", 
                        "Количество продутка", 
                        "Брак продукта",
                        "Количество ингридиентов",
                        "Цена ингридиентов",
                        "Количество заказов"])
```

```python
"""Оптовая цена на продукты в виде точечной диаграмы"""
answer_df.drop_duplicates(subset="name_product").plot.scatter(x = 'name_product', y = 'wholesale_price_product', figsize = (15, 3))
```
